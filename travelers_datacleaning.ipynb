```python
#import data

import pandas as pd

file_path = '/Users/ianpoe/Downloads/2024-travelers-university-modeling-competition/train_data.csv'
df = pd.read_csv(file_path)
df.head()
    

#make channel binary
df['channel'] = df['channel'].apply(lambda x: 1 if x == 'Retail' else 0)

#make trm_len_mo binary
df['trm_len_mo'] = df['trm_len_mo'].apply(lambda x: 1 if x == 12 else 0)


#one-hot encoding to convert categorical into binary
df = pd.get_dummies(df, columns=['acq_method', 'bi_limit_group', 'geo_group', 
                                 'household_group', 'pay_type_code', 'pol_edeliv_ind',
                                 'prdct_sbtyp_grp', 'product_sbtyp', 'telematics_ind'], 
                    drop_first=True, dtype=int)

df.head()
```

```python
#export cleaned csv
output_name = '/Users/ianpoe/Downloads/2024-travelers-university-modeling-competition/cleaned_training_data.csv'
df.to_csv(output_name, index=False)
```
